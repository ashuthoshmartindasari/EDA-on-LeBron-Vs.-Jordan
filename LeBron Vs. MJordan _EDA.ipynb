## Import all necessary packages.

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
from numpy.random import randn

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Importing Dataset Lebron

```python
data_lebron = pd.read_csv('C:/Users/Administrator/Documents/NBA/nba-players-career-game-log/lebron_career.csv')
data_lebron = data_lebron.set_index('game')
data_lebron.head()
```

```python
data_lebron['result'] = data_lebron['result'].str[:1]
```

```python
data_lebron.columns = map(str.upper, data_lebron.columns)          #Columns to Upper case
```

```python
data_lebron.head()
```

## Lebron James Career points against each Team

```python
fig, ax = plt.subplots(figsize=(16,7))

data_lebron_opp = data_lebron.pivot_table(columns='RESULT',index='OPP',values='PTS')
data_lebron_opp.plot(ax=ax, kind='bar')

ax.set_title("Lebron Career Points Against Each Team", fontsize=14)
ax.legend(loc='upper right', title='Game Result')
plt.show()
```

```python
data_lebron['AGE'] = data_lebron['AGE'].astype(str)
data_lebron['AGE'] = data_lebron['AGE'].str.replace('-','.')
data_lebron['AGE'] = data_lebron['AGE'].astype(float)
```

## Age Vs. Points (Data--> Lebron James)

```python
sns.lmplot("AGE","PTS", data=data_lebron, order=2, line_kws={'color':'indianred', 'linewidth':1})
```

## import Jordan Dataset

```python
data_jordan = pd.read_csv('C:/Users/Administrator/Documents/NBA/nba-players-career-game-log/jordan_career.csv')
data_jordan = data_jordan.set_index('game')
data_jordan.columns = map(str.upper, data_lebron.columns)
data_jordan.head()
```

```python
data_jordan['AGE'] = data_jordan['AGE'].astype(str)
data_jordan['AGE'] = data_jordan['AGE'].str.replace('-','.')
data_jordan['AGE'] = data_jordan['AGE'].astype(float)
```

```python
data_jordan['RESULT'] = data_jordan['RESULT'].str[:1]
data_jordan_opp = data_jordan.pivot_table(columns='RESULT',index='OPP',values='PTS')
```

```python
data_jordan.head(3)
```

```python
ser1 = data_lebron.mean()
ser2 = data_jordan.mean()
lebron_mean = pd.DataFrame(ser1).transpose()
jordan_mean = pd.DataFrame(ser2).transpose()
dataset = pd.DataFrame(pd.concat([lebron_mean, jordan_mean], ignore_index=True))
dataset = dataset.rename(index={0:'Lebron',1:'Jordan'})
dataset.head(3)
```

## Plotting a Radar Chart

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, polar=True)

labels = np.array(['MP', 'FG', 'FGP', 'FTA', 'TRB', 'AST', 'STL', 'PTS','BLK'])
stats1=dataset.loc['Lebron',labels].values
stats2=dataset.loc['Jordan',labels].values

angles=np.linspace(0, 2*np.pi, len(labels)+1, endpoint=False)
stats1=np.concatenate((stats1,[stats1[0]]))
stats2=np.concatenate((stats2,[stats2[0]]))

ax.plot(angles, stats1, 'o-', linewidth=2)
ax.fill(angles, stats1, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)

ax.plot(angles, stats2, 'o-', linewidth=2)
ax.fill(angles, stats2, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)

ax.set_title("Lebron Vs. Jordan Career Average Overview")
ax.grid(True)

label_texts=('Lebron','Jordan')
ax.legend(label_texts, loc='upper right', bbox_to_anchor=(0.03 , 0.01))
```

## Comparing Career Avg. of Lebron and Jordan

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(5, 14))
fig.subplots_adjust(wspace=0.5)

ax1.set_title("Jordan Career Average.")
sns.heatmap(data_jordan_opp, ax=ax1, annot=True, cmap="rainbow",center=25, linewidths=.1, cbar=False, robust=True)

sns.heatmap(data_lebron_opp, ax=ax2, center=25, cmap="rainbow", vmin=15, vmax=35, annot=True,cbar=True, linewidth=.1)
ax2.set_title("Lebron Career Average.")
```
